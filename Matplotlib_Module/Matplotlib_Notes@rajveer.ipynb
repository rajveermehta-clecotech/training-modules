### # Pyplot - Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
#or plt.show() method apart from jupyter notebook.
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#Functional method for plotting graphs

plt.plot(x,y)
#plt.show() if not in jupyter notebok.
```

```python
# adding x label, y label, title to plot

plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

```python
# plotting two subplot on same canvas
 
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(x,y,"b")
```

```python
# OOPs method for plotting graph

fig = plt.figure()

axes =  fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('y Label')
axes.set_title('Set title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('first plot')

axes[1].plot(y,x)
axes[1].set_title('second plot')

plt.tight_layout()
```

### # Figure size & DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

### # Adding a legend

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label="X label")
ax.plot(x,x**3, label="Y labels")

ax.legend() #we can pass in loc parameter to shift legend
```

### # Plot Apperance

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color="red",linewidth=10) #linewidth or lw parameter
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color="red",linewidth=1,linestyle="--") #linestyle or ls parameter
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color="red",linewidth=1,linestyle="--", marker="o") #marker parameter
```
